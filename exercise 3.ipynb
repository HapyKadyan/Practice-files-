```python
def prime(x, y):
    prime_list = []
    for i in range(x, y):
        if i == 0 or i == 1:
            continue
        else:
            for j in range(2, int(i/2)+1):
                if i % j == 0:
                    break
            else:
                    prime_list.append(i)
    return prime_list
 
# Driver program
starting_range = 1
ending_range = 300
lst = prime(starting_range, ending_range)
if len(lst) == 0:
    print("There are no prime numbers in this range")
else:
    print("The prime numbers in this range are: ", lst)
```

3.	WAP which repeatedly reads numbers until the user enters “done”. Once “done” is entered, print out the total, count, and average of the numbers. Enter a number: 4

Invalid input
Enter a number: 7

16	3	5.33333333333


```python
total=0
count=0
print("enter only numberic value and enter done to stop ")
while True:
    try:
        n=input("enter a number: ")
        if n.lower()=="done":
            break 
        else:
            total=total+int(n)
            count=count+1
            
    except Exception as e:
        print("wrong value enter again")
print("total is ",total) 
print("count is ",count)
print("average is ",total/count)
```

8.	Recall the problem when you kept prompting the user to withdraw money from the wallet to start the game. Integrate this program with the one you have written for problem 7.

```python
from random import randint
a=randint(0,9)
print(a)
balance=eval(input("enter your balance: "))
c=input("enter yes or no to play this game: ")
b=c.lower()
if b.lower()=="yes":
    if balance>=500:
        print("game has been started try your luck")
        for i in range(1,4):
            num=eval(input("Guess the number: "))
            if i==1 and a==num:
                print("you have just won Rs 5,000 Your balance is Rs {}".format(balance+5000))
                break
            elif a==num:
                print("you won just won Rs 1,000 Your balance is Rs {}".format(balance+1000))
                break
            elif i==3:
                print("you lost Your balance is rs {}".format(balance-500))
            else:
                print("try again")
    else :
        print("you don't have enough money ")
        print("bring rs {} if you wanna try your luck".format(500-balance))
    
else:
    print("have a wonderful day!")
```
