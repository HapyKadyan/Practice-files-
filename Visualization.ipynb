#### package name: matplotlib (matplotlibrary) 

```python
import matplotlib.pyplot as plt

# numpy  pandas   matplotlib  default no need to install
```

- plt.plot() : Create a plot object
    
- plt.title(): Add a title for plot object
    
- plt.xlabel(): specify lables of x-axis
    
- plt.ylabel(): specify labels of y-axis
    
- plt.show(): Display the plot

**line plot**

```python
# Suppose You want to plot a sales plot
# Sales happens in every month
import numpy as np
import matplotlib.pyplot as plt
sales=[100,112,203,304,405,506,96,700,600,80,900,1000]
months=np.arange(1,13)
```

```python
plt.plot(months,sales,color='b',marker='*')
plt.xlabel("Months")
plt.ylabel("Sales")
plt.title("Sales view")
plt.savefig("Months_Sales.jpg")
plt.show()
```

```python
plt.plot(months,sales,color='b',marker='*')
plt.xlabel("Months")
plt.ylabel("Sales")
plt.title("Sales view")
plt.grid()
plt.show()
```

```python
plt.plot(months,sales,color='b',marker='*')
plt.xlabel("Months")
plt.ylabel("Sales")
plt.title("Sales view")
plt.grid(linestyle='-.',linewidth=1,color='green')
plt.show()
```

```python
#M-1:
x=np.arange(-6,7)
y=[i**2 for i in x]
plt.plot(x,y,color='b',marker='*')
plt.xlabel("X")
plt.ylabel("Y")
plt.title("Parabola")
plt.grid(linestyle='-.',linewidth=1,color='green')
plt.show()
```
