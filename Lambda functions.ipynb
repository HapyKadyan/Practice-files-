$map$

- map means maping 
- when we have to do operation like +,-,*,/ 
- appending elements in list or tuple

```python
list1=[1,2,3,4,5]
list2=list(map(lambda x: x*2,list1))
```

```python
list2
```

```python
map(lambda_expression,sequence) # map,filter,reduce takes two arguments (lambda_expression,sequence)
```

```python
strings=['winter','is','coming']
#['WINTER!!','IS!!','COMING!!']
string=list(map(lambda x:x.upper()+'!!',strings))
string
```

```python
num1=[1,2,3]
num2=[10,20,30]
add= list(map(lambda x,y:x+y>30,num1,num2))
add
```

```python
list3=[]
for i,j in zip(num1,num2):
    add=i+j
    if add>30:
        list3.append(i+j)
list3
```

```python
list1=['Apple','banana','car','bat','dog']
# I need items those length greater than 3
# ['Apple','banana']
```

```python
items=[]
for i in list1:
    if len(i)>3:
        items.append(i)

items
```

$filter$

- filter means filtering 
- when we have to operation like <,> %,in(operator)

filter takes two arguments (lambda expression,sequence)

- if we are using map,filter,reduce 
- takes two arguments
- and list or tuple must be used 

```python
li=[87,87,90,65,45,32]
newli=list(filter(lambda x: (x%2!=0),li)) # append elements in list or tuple
newli
```

```python
li=[87,87,90,65,45,32]
newli=list(filter(lambda x: (x%2!=0),li))
newli
```
