## np.where

```python
condition=bank_data["loan"]=="no"
# whenever this condition is true replace with 0 i.e; no is replaced by 0
# whenever this condition is not true replace with 1 i.e; here other then no values are replaced by 1
# other than no , only one label is there i.e; yes 
df=np.where(condition,0,1)
# first value(zero) is for true condition 
# second value(one) is for false condition 
```

```python
df
```

```python
bank_data
```

```python
bank_data["loan"]=np.where(condition,0,1)
```

```python
bank_data
```

```python
# other way but condition inside 
bank_data["loan"]=np.where(bank_data["loan"]=="no",0,1)
```

```python
bank_data["loan_new"]=np.where(bank_data["loan"]=="no",0,1)
# adding new column loan_new 
```

## label encoder

- it will provide the labels automatically 
- suppose you have two labels yes and no, then yes=0 and no=1
- if you have 3 labels such as singlr,married,divorce then single=0,married=1,divorce=2
- manual map is not required 
- suppose there are 100 labels,automatically it will form 0 to 99
- best for ordinal values ex: primary,secondary,teritory
