```python
dict1={"Names":['Anjali','Balaram','Chinni','Diya'],
      "Age":[20,25,24,32],
      "City":['Kolkata','Delhi','Vizag','Noida']}

df3=pd.DataFrame(dict1,columns=['Names','Age','City'])
df3
```

```python
# if you have multiple dataframe append
```

```python
df1.append(df3)
# after anil index start from 0
# to avoid this use ignore_index
```

```python
df1.append(df3,ignore_index=True)
```

```python
df1.append(df2)
# df1 have 3 columns 
# df2 have 2 columns
# full outer join
# Names age city          city name
# names age city name
```

```python
df1.append(df2,ignore_index=True)
```

```python
pd.concat([df1,df3]) # append
# values are appending based on rows
pd.concat([df1,df3],axis=0) # both are same axis=0 default
```

```python
pd.concat([df1,df3],ignore_index=True) 
```

```python
pd.concat([df1,df3],axis=1)
# datasets concatenate based on columns
```
