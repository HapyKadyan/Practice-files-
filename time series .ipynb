```python
a=df_daily
b=a.groupby(["Year"]).sum().sort_values("Year",ascending=True).style.background_gradient(cmap='Spectral_r')
b
```

```python
a=df_hour
b=a.groupby(["Hour"]).sum().sort_values("Hour",ascending=True).style.background_gradient(cmap='Purples')
b
```

```python
a=df_hour
b=a.groupby(["Weekday Name"]).sum().sort_values("Weekday Name",ascending=True).style.background_gradient(cmap='twilight_shifted')
b
```

```python
a=df_hour
b=a.groupby(["Month"]).sum().sort_values("Month",ascending=True).style.background_gradient(cmap="gist_yarg")
b
```

```python
col=df_hour.columns[1:9]
plt.figure(figsize=(10,6))
for i in col:
    sns.lineplot(y=df_hour[i],x=df_hour["Weekday Name"],ci=None,marker="o",label=i)
plt.xlabel("Weekdays")
plt.ylabel("Sales")    
plt.legend()    
plt.show()
```
