**np.arange()**

- np.arange() --- also used to create numpy array 
- the numbers generated with some difference

```python
# [i for i in range(2,20,2)]
even=np.arange(start=2,stop=20,step=2)
print(even)
```

```python
even=np.arange(start=2,stop=20,step=2,dtype=int)
print(even)
```

```python
even=np.arange(start=2,stop=20,step=2,dtype=float)
print(even)
```
