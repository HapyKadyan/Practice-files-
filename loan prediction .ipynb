```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
loan_data=pd.read_csv("D:\\data\\loan.csv")
loan_data.head()
```

```python
loan_data.shape
```

```python
loan_data.info()
```

```python
loan_data.drop("Loan_ID",axis=1,inplace=True)
# we are droping loan id 
```

```python
loan_data.head()
```

```python
# in gender column we have 13 missing value 
# so we droping them 
nw=loan_data["Gender"]=loan_data["Gender"].dropna()
nw.shape
# instead of droping fill the values with mode 
```

```python
loan_data["Gender"]=loan_data["Gender"].fillna(loan_data["Gender"].mode())
loan_data["Gender"].shape
```

```python
loan_data.isnull().sum()
```

```python
# now fill married 
loan_data["Married"].fillna(loan_data["Married"].mode(),inplace=True)
loan_data["Married"].isna().sum()
```

```python
# dependents 
loan_data["Dependents"].fillna(loan_data["Dependents"].mode(),inplace=True)
```

```python
# self_employed 
loan_data["Self_Employed"].fillna(loan_data["Self_Employed"].mode(),inplace=True)
```

```python
# Loan amount 
# how can i check if there are outliers or not ?
loan_data["LoanAmount"]=loan_data["LoanAmount"].fillna(loan_data["LoanAmount"].mean())
```

```python
loan_data.describe()
# in coapplicantincome there may be a outliers 
```

```python
# now Loan_Amount_Term
loan_data["Loan_Amount_Term"]=loan_data["Loan_Amount_Term"].fillna(loan_data["Loan_Amount_Term"].mean())
```

```python
# Credit_History
loan_data["Credit_History"]=loan_data["Credit_History"].fillna(loan_data["Credit_History"].mean())
```
