```python
def handlesalary(value):
    if value<50000:
        return "less than 50k"
    elif value<80000:
        return "50k-80k"
    elif value<100000:
        return "80k-1L"
    elif value<300000:
        return "1L-3L"
    elif value<500000:
        return "3L-5L"
    elif value<800000:
        return "5L-8L"
    elif value<1000000:
        return "8L-10L"
    elif value<1500000:
        return "10L-15L"
    elif value<2000000:
        return "15L-20L"
    elif value<3000000:
        return "20L-30L"
    else:
        return "above 30L"
```

```python
df["Salary"]=df["Salary"].apply(handlesalary)
```

```python
import plotly.graph_objs as go
plt.figure(figsize=(4,3))
labels=["Tier1","Tier2"]
values = df['CollegeTier'].value_counts()
colors = ['mediumslateblue', 'darkorange']
fig = go.Figure(data=[go.Pie(labels=labels,
                             values=values,hole=.3)])
fig.update_traces(hoverinfo='label+percent', textinfo='value', textfont_size=20,
                  marker=dict(colors=colors, line=dict(color='#000000', width=3)))
fig.update_layout(title="College Tier",
                  titlefont={'size': 30},      
                  )
fig.show()
```
